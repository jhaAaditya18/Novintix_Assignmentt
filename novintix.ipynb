**Task 2 — Supervised Learning (Time-Series Forecasting for ALL 4 Variables)**

**Reasoning**:
The subtask requires creating a list of specified target variables. I will create a Python code block to define this list.



```python
target_variables = [
    'LV ActivePower (kW)',
    'Wind Speed (m/s)',
    'Theoretical_Power_Curve (KWh)',
    'Wind Direction (°)'
]
print("Target variables defined:", target_variables)
```

**Reasoning**:
I will create lagged features for each target variable as described in the subtask. This involves iterating through the target variables and applying a shift operation to generate lagged columns for a window size of 5.



```python
window_size = 5

for var in target_variables:
    for i in range(1, window_size + 1):
        df[f'{var}_lag_{i}'] = df[var].shift(i)

# Drop rows with NaN values resulting from the lagging process
df.dropna(inplace=True)

print(f"DataFrame shape after creating {window_size} lagged features and dropping NaN values: {df.shape}")
print("First 5 rows of the DataFrame with lagged features:")
print(df.head())
```

**Reasoning**:
I will proceed with splitting the data into features (X) and targets (y), then chronologically divide them into training and testing sets. After that, I will apply `MinMaxScaler` to scale both features and targets separately. Finally, I will print the first 5 rows of the scaled training and testing sets to verify the transformation.



```python
from sklearn.preprocessing import MinMaxScaler

# 1. Define X (features) and y (target)
X = df.filter(like='_lag_', axis=1)
y = df[target_variables]

# 2. Calculate the split point for an 80/20 chronological split
split_point = int(len(df) * 0.8)

# Split data chronologically
X_train = X.iloc[:split_point]
X_test = X.iloc[split_point:]
y_train = y.iloc[:split_point]
y_test = y.iloc[split_point:]

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")

# 3. Initialize separate MinMaxScaler objects
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

# 4. Fit scalers to training data
feature_scaler.fit(X_train)
target_scaler.fit(y_train)

# 5. Transform all datasets using their respective fitted scalers
X_train_scaled = feature_scaler.transform(X_train)
X_test_scaled = feature_scaler.transform(X_test)
y_train_scaled = target_scaler.transform(y_train)
y_test_scaled = target_scaler.transform(y_test)

# Convert scaled arrays back to DataFrames for easier viewing, maintaining column names
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)
y_train_scaled = pd.DataFrame(y_train_scaled, columns=y_train.columns, index=y_train.index)
y_test_scaled = pd.DataFrame(y_test_scaled, columns=y_test.columns, index=y_test.index)

# 6. Print the first 5 rows of the processed training and testing sets
print("\nX_train_scaled (first 5 rows):\n", X_train_scaled.head())
print("\nX_test_scaled (first 5 rows):\n", X_test_scaled.head())
print("\ny_train_scaled (first 5 rows):\n", y_train_scaled.head())
print("\ny_test_scaled (first 5 rows):\n", y_test_scaled.head())
```

**Reasoning**:
I will import the `RandomForestRegressor` model, instantiate it, and then fit it to the scaled training data as specified in the instructions to train the time-series forecasting model.



```python
from sklearn.ensemble import RandomForestRegressor

# Instantiate the RandomForestRegressor model
# Using random_state for reproducibility
model = RandomForestRegressor(random_state=42)
print("RandomForestRegressor model instantiated.")

# Fit the model to the scaled training data
model.fit(X_train_scaled, y_train_scaled)
print("Model training completed successfully.")
```

**Reasoning**:
To evaluate the model, I need to make predictions on the test set, inverse transform both predictions and actual values, and then calculate RMSE and MAE for each target variable using scikit-learn metrics.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# 1. Use the trained model to make predictions on X_test_scaled
y_pred_scaled = model.predict(X_test_scaled)

# Convert y_pred_scaled to a DataFrame with appropriate columns and index
y_pred_scaled_df = pd.DataFrame(y_pred_scaled, columns=target_variables, index=X_test_scaled.index)

# 2. Inverse transform y_pred_scaled and y_test_scaled
y_pred = target_scaler.inverse_transform(y_pred_scaled_df)
y_test_actual = target_scaler.inverse_transform(y_test_scaled)

# Convert to DataFrames for easier column access and alignment
y_pred = pd.DataFrame(y_pred, columns=target_variables, index=X_test_scaled.index)
y_test_actual = pd.DataFrame(y_test_actual, columns=target_variables, index=X_test_scaled.index)

print("Model predictions generated and inverse-transformed.")

# 4. For each target variable, calculate RMSE and MAE
print("\n--- Model Performance Evaluation ---")
for var in target_variables:
    rmse = np.sqrt(mean_squared_error(y_test_actual[var], y_pred[var]))
    mae = mean_absolute_error(y_test_actual[var], y_pred[var])
    print(f"\nMetric for {var}:")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE: {mae:.4f}")

print("------------------------------------")
```

```python
import matplotlib.pyplot as plt

# 1. For each target_variable, create a new figure for plotting.
for var in target_variables:
    plt.figure(figsize=(15, 6))

    # 2. Plot actual and predicted values
    plt.plot(y_test_actual.index, y_test_actual[var], label='Actual', color='blue')
    plt.plot(y_pred.index, y_pred[var], label='Predicted', color='red', linestyle='--')

    # 3. Add appropriate x-axis and y-axis labels
    plt.xlabel('Date/Time')
    plt.ylabel(var)

    # 4. Add a title to each plot
    plt.title(f'{var} - Actual vs. Predicted')

    # 5. Include a legend
    plt.legend()

    # 6. Display each plot
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
To calculate the power deviation, I will create a new column named 'Power Deviation' in the DataFrame by subtracting 'Theoretical_Power_Curve (KWh)' from 'LV ActivePower (kW)'.



```python
df['Power Deviation'] = df['LV ActivePower (kW)'] - df['Theoretical_Power_Curve (KWh)']

print("DataFrame with 'Power Deviation' column:")
print(df.head())
```

**Reasoning**:
I will now apply an Isolation Forest model to the 'Power Deviation' column to detect anomalies. The model will be fit on this single feature to identify instances that significantly deviate from the norm.



```python
from sklearn.ensemble import IsolationForest

# Reshape the 'Power Deviation' column for the Isolation Forest model
X_anomaly = df['Power Deviation'].values.reshape(-1, 1)

# Initialize and train the Isolation Forest model
# contamination is the proportion of outliers in the dataset, often set based on domain knowledge or tried values.
# A common practice is to start with a small value like 0.01 or 0.05.
model_if = IsolationForest(contamination=0.05, random_state=42)
model_if.fit(X_anomaly)

# Predict anomalies (-1 for outliers, 1 for inliers)
df['Anomaly'] = model_if.predict(X_anomaly)

print("Isolation Forest model trained and anomalies detected.")
print("Number of anomalies detected: ", df[df['Anomaly'] == -1].shape[0])
print("First 5 rows with Anomaly column:")
print(df[['Power Deviation', 'Anomaly']].head())
```

**Reasoning**:
To calculate the performance ratio and handle potential division by zero or NaN values, I will perform the division, convert infinities to NaN, and then fill NaN values with 0.0, as per the instructions.



```python
df['Performance Ratio'] = df['LV ActivePower (kW)'] / df['Theoretical_Power_Curve (KWh)']

# Convert any infinite values (resulting from division by zero) to NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill any NaN values in 'Performance Ratio' with 0.0
df['Performance Ratio'] = df['Performance Ratio'].fillna(0.0)

print("DataFrame with 'Performance Ratio' column:")
print(df[['LV ActivePower (kW)', 'Theoretical_Power_Curve (KWh)', 'Power Deviation', 'Performance Ratio']].head())
```

**Reasoning**:
To scale the 'Performance Ratio' to a range of 0 to 100, I will use `MinMaxScaler` from `sklearn.preprocessing`, instantiate it with the specified range, reshape the data, and then apply fit_transform to create the new 'Performance Score' column.



```python
from sklearn.preprocessing import MinMaxScaler

# Reshape the 'Performance Ratio' column for scaling
performance_ratio_reshaped = df['Performance Ratio'].values.reshape(-1, 1)

# Instantiate MinMaxScaler with the desired feature range
scaler = MinMaxScaler(feature_range=(0, 100))

# Fit and transform the data to create the 'Performance Score'
df['Performance Score'] = scaler.fit_transform(performance_ratio_reshaped)

print("DataFrame with 'Performance Ratio' and 'Performance Score' columns:")
print(df[['Performance Ratio', 'Performance Score']].head())
```

**Reasoning**:
To categorize the 'Performance Score' into 'Good', 'Moderate', or 'Poor' states, I will define thresholds and use `np.select` to create a new column 'Performance State' based on these conditions.



```python
import numpy as np

# Define the thresholds for categorization
# These thresholds can be adjusted based on domain expertise
poor_threshold = 20
moderate_threshold = 70

# Define conditions for categorization
conditions = [
    (df['Performance Score'] < poor_threshold),
    (df['Performance Score'] >= poor_threshold) & (df['Performance Score'] < moderate_threshold),
    (df['Performance Score'] >= moderate_threshold)
]

# Define the corresponding choices
choices = ['Poor', 'Moderate', 'Good']

# Apply the categorization
df['Performance State'] = np.select(conditions, choices, default='Unknown')

print("DataFrame with 'Performance State' column:")
print(df[['Performance Score', 'Performance State']].head())
print("\nValue counts for 'Performance State':")
print(df['Performance State'].value_counts())
```

**Reasoning**:
To generate automated suggestions for each turbine state, I will define a dictionary that maps each 'Performance State' to a specific suggestion and then apply this mapping to create a new 'Suggestion' column.



```python
suggestions_map = {
    'Poor': 'Check for maintenance, investigate operational issues.',
    'Moderate': 'Monitor performance closely, consider minor adjustments.',
    'Good': 'Operating optimally, continue routine checks.'
}

df['Suggestion'] = df['Performance State'].map(suggestions_map)

print("DataFrame with 'Performance State' and 'Suggestion' columns:")
print(df[['Performance State', 'Suggestion']].head())
print("\nValue counts for 'Suggestion':")
print(df['Suggestion'].value_counts())
```

**Reasoning**:
To define the hyperparameter search space as instructed, I will create a Python dictionary named `param_grid` containing lists of values for `n_estimators`, `max_depth`, and `min_samples_leaf`.



```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_leaf': [1, 2, 4]
}

print("Hyperparameter grid defined:")
print(param_grid)
```

**Reasoning**:
To perform Grid Search Cross-Validation as instructed, I will import `GridSearchCV`, instantiate `RandomForestRegressor`, set up `GridSearchCV` with the provided parameters, fit it to the scaled training data, and then print the best parameters and score.



```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# 1. Create an instance of RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)

# 2. Instantiate GridSearchCV
# param_grid was defined in a previous step
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=3,  # 3-fold cross-validation
    scoring='neg_mean_squared_error', # Optimize for lower RMSE
    n_jobs=-1  # Use all available CPU cores
)

print("Starting GridSearchCV...")
# 3. Fit the GridSearchCV object to your scaled training data
grid_search.fit(X_train_scaled, y_train_scaled)
print("GridSearchCV completed.")

# 4. Print the best parameters found by the grid search
print("\nBest parameters found:", grid_search.best_params_)

# 5. Print the best score (negative mean squared error) found by the grid search
print("Best score (negative mean squared error):", grid_search.best_score_)
```
